# Box plot

```python
import matplotlib.pyplot as plt
import numpy as np
x=[10,20,30,40,50,60,70,190]
plt.boxplot(x,vert=True,patch_artist=True,labels=["box plot"],sym="r+",boxprops=dict(color="r"),capprops=dict(color="g"))
plt.show()
```

# stack plot

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5,6]
area1=[1,2,3,4,5,6]
area2=[2,4,6,8,10,12]
area3=[3,6,9,12,15,18]
list=[area1,area2,area3]
plt.stackplot(x,area1,area2,area3,labels=list)
plt.legend(loc=2)
plt.show()
```

# savefig

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.subplot(2,2,1)
plt.plot(x,y,color="r")
plt.show()
plt.savefig("fig1")

x1=[1,2,3,4,5]
y2=[1,2,3,4,5]
plt.subplot(2,2,1)
plt.plot(x1,y2,color="g")
plt.show()
plt.savefig("fig2")
```
