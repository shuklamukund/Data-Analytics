```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('automotive')
dataset.head()
```

```python
dataset['price'].fillna(dataset['price'].mean(),inplace=True)
```

```python
dataset.describe()
```

```python
sns.distplot(dataset['price'])
plt.show()
```
