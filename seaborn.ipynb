```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var1=[1,2,3,4,5,6,7]
var2=[2,4,6,8,10,12,14]

x=pd.DataFrame({"var_1":var1,"var_2":var2})

sns.lineplot(x="var_1",y="var_2",data=x)
plt.show()
```
