# one hot encoding

```python
import pandas as pd
```

```python
dataset=pd.read_csv('automotive')
dataset
```

```python
dataset.isnull().sum()
```

```python
en_data=dataset[['num-of-doors']]
en_data
```

```python
en_data.info()
```

```python
pd.get_dummies(en_data).shape
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first')
```

```python
ar=ohe.fit_transform(en_data).toarray()
ar
```

```python
new_dataset=pd.DataFrame(ar,columns=["num-of-doors_four","general"])
new_dataset
```

```python
print(new_dataset.columns)
```

```python
new_dataset.drop(columns=['general'],inplace=True)
print(new_dataset)
```

# ordinal encoding

```python
df=pd.DataFrame({"size":["s","m","l","xl","s","s","l","xl","m"]})
df.head()
```

```python
ord_data=[['s','m','l','xl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ord_data)
df[['ord_enc']]=oe.fit_transform(df[['size']])
```
